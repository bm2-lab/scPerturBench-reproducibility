```python
# 自定义渐变
colors = ['darkgreen', '#ffffff', '#9B0056']
custom_cmap = LinearSegmentedColormap.from_list("science", colors)
custom_cmap
```

```python
for datasets_name in cellular_datasets_name:
    cellular_datasets_path = cellular_datasets_dir + '/' + datasets_name
    print(cellular_datasets_path)
    pkl_path = cellular_datasets_path + '/outSample_Sim_pca.pkl'
    if os.path.exists(pkl_path):
        with open(pkl_path, 'rb') as f:
            mydict = pickle.load(f)
        data = 1 - mydict['pearson_distance']
        #### heatmap ####
        # 1. 绘制 clustermap
        g = sns.clustermap(
            data,
            fmt=".1f",
            annot=True, # 显示数值
            figsize=(8, 8), 
            annot_kws={"size": 12}, # 热图数值字体大小
            vmin=-1, # 最小值
            vmax=1, # 最大值
            cmap=custom_cmap, 
            cbar_kws={
                "label": "Pearson Correlation",
                "ticks": [-1, 0, 1]
            }
        )

        # 2. 调整 x/y 轴标签
        g.ax_heatmap.set_xticklabels(
            g.ax_heatmap.get_xticklabels(),
            fontsize=12,
            rotation=45,
            ha="right"  # 标签对齐方式
        )
        g.ax_heatmap.set_yticklabels(
            g.ax_heatmap.get_yticklabels(),
            fontsize=12
        )

        # 3. 调整 x 轴和 y 轴标题
        g.ax_heatmap.set_xlabel("", fontsize=1, labelpad=1)  # labelpad 控制标题与轴的距离
        g.ax_heatmap.set_ylabel("", fontsize=1, labelpad=1)

        # 4. 调整 colorbar 标题
        g.cax.set_ylabel("Pearson Correlation", fontsize=10, rotation=90, va="top")
        # plt.show()

        # 5. 保存文件
        plt.savefig(datasets_name + "_clustermap.png", dpi=300, bbox_inches="tight")


        #### umap ####
        h5ad_path = cellular_datasets_path + '/UmapAnalysis/filter_hvg5000_logNor_control.h5ad'

        adata = sc.read_h5ad(h5ad_path)
        sc.pl.umap(adata, color = 'condition1', show = False)
        plt.savefig(datasets_name + "_umap.png", dpi=300, bbox_inches="tight")
    else:
        print(f"路径 '{pkl_path}' 不存在")
```
